```python
from PIL import Image
from pylab import *
```

```python
# 读取图像
pil_im = Image.open('demo.jpg')
# pil_im = pil_im.resize((128,300))
# 将图像读取到数组中
im = array(pil_im)
im.shape
# 绘制图像
imshow(im)
# 一些点
x = [100,100,400,400]
y = [200,500,200,500]
# 使用红色星状标记绘制点
plot(x,y,'r*')# 绘制四个点(100,200)、(100,500)、(400,200)、(400,500)
plot(x[:2],y[:2])# 由(100,200)、(100,500)这两个点连接形成的直线
# 显示绘制的图像
title('demo.jpg')
axis('off')
show()
```

```python
# plot((100,100),(200,100))
x = (100,100)
y = (200,100)
plot(x,y,'ks:')
show()
```

```python
# 图像轮廓图
im = array(Image.open('demo.jpg').convert('L'))
print('im.shape',im.shape)
# 新建一个图像
figure()
# 不使用颜色信息
gray()
# 在原点的左上角显示轮廓线
contour(im,origin='image')
axis('equal')
axis('off')
show()
```

```python
# 图像直方图
figure()
# 因为 hist() 只接受一维数组作为输入，所以我们在绘制图像直方图之前，必须先对图像进行压平处理。
# flatten() 方法将任意数组按照行优先准则转换成一维数组
# im的shape为(359, 252)
hist(im.flatten(),128)# 第二个参数指定小区间的数目
show()
```
