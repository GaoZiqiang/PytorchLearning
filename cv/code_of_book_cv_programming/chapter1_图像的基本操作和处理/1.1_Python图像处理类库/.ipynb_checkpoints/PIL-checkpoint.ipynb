```python
from PIL import Image
```

```python
# 读取图像
pil_im = Image.open('demo.jpg')
# 创建最长边为 128 像素的缩略图
pil_im.thumbnail((128,128))
pil_im
```

```python
# 图像裁剪
box = (200,100,800,400)
region = pil_im.crop(box)# 裁剪
region = region.transpose(Image.ROTATE_180)# 对裁剪后的region旋转
pil_im.paste(region,box)# 将裁剪后的region重新贴回原图像
pil_im
out = pil_im.rotate(45)# 旋转
out = pil_im.resize((128,128))# risize
out
```

```python
# 对图像进行颜色转换
pil_im.convert('L')# 转换成灰度图像
```
