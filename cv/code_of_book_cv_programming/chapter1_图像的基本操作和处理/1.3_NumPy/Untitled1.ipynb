```python
from PIL import Image
from numpy import *
from pylab import *
```

```python
im = array(Image.open('demo.jpg'))
im.shape
# im.dtype
# 数组中的元素可以使用下标访问
# 于坐标 i、j，以及颜色通道 k 的像素值
value = im[0,0,1]
value
im[:,0] = 100
im[:,0,0]
im[:100:50].sum()
im[0,:,2].mean()
```

```python
im1 = array(Image.open('demo.jpg').convert('L'),'f')
im2 = 255 - im1# 对图像进行反相处理
im3 = (100.0/255)*im1 + 100# 将图像像素值变换到100~200之间
im4 = 255.0*(im1/255.0)**2# 对图像像素值求平方
int(im1.min())
int(im1.max())
imshow(im2)
imshow(im3)
imshow(im4)
axis('off')
# show()
im1.dtype
```

```python
# array()变换的相反操作
pil_im = Image.fromarray(uint8(im1))# 对于不确定的数据类型，在进行反向变换之前要将array转换为uint8型，不然会出错
pil_im
```
