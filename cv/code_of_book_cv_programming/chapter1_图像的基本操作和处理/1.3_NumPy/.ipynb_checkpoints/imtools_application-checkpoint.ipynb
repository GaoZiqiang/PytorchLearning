```python
from PIL import Image
from numpy import *# import方式有两种，第一种import ** as @@，在调用**的属性和方法时需要使用点标记法引用
#from numpy import *# 使用from ** import *这种方法就不需要使用点标记法显式引用了
import imtools
from pylab import *
```

```python
#im = array(Image.open('demo.jpg'))
pil_im = Image.open('demo.jpg')
im = array(Image.open('demo.jpg').convert('L'))
im
```

```python
# 进行直方图均衡化
im2,cdf = imtools.histeq(im)
# 画图，展示图像直方图
figure()
hist(im.flatten(),128,color='r')# 原图的直方图
hist(im2.flatten(),128)# 经均衡化后的直方图，可以看到，分布更加均匀
show()
# 看一下cdf
cdf.shape
```

```python
# 直方图均衡化之前的图像
pil_im
```

```python
# 直方图均衡化之后的图像，可以看到，原先图像灰色区域的细节变得清晰
pil_im2 = Image.fromarray(uint8(im2))
pil_im2
```

```python
# 画cdf累积分布函数
figure()
x = arange(0,256)
y = arange(0,256)
plot(x,cdf)
title('Cumulative Distribution Function')
show()
```
